# Defining Suits, Ranks And Values(Powers) for the cards in deck
### Also importing necessary module

```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

# Card Class with attributes: Rank, Suit and Value

```python
class Card:
    def __init__(self, rank, suit):
        self.rank = rank
        self.suit = suit
        self.value = values[rank]

    def __str__(self):
        return f"{self.rank} of {self.suit}"
```

# Deck Class

```python
class Deck:

    def __init__(self):
        self.all_cards = []
        for suit in suits:
            for rank in ranks:
                created_card = Card(rank,suit)
                self.all_cards.append(created_card)
    
    def __str__(self):
        # To print the deck as a list
        
        return f"{self.all_cards}"
    
    def shuffle(self):
        # To Shuffle the contents of deck
        
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        # To return the topmost card in the deck
        
        return self.all_cards.pop(0)    
```

```python
class Player:
    
    def __init__(self, name):
        self.name = name
        self.all_card = []
        
    def add_card(self, card):
        # To add card in the player's deck at bottom
        
        if type(card) == type([]):
            self.all_card.extend(card)
        else:
            self.all_card.append(card)
    
    def remove_card(self):
        # To return the player's topmost card and remove it from the player's deck at the same time
        
        return self.all_card.pop(0)
    
    def __str__(self):
        # To return the count of cards left in player's deck
        
        return f"{self.name} - {len(self.all_card)} cards."
    
    
```

```python
def play_warofcards():
    # To keep check of Rounds
    
    round = 0
    
    # To decide when to end the game
    
    game_on = True


    player_a = Player("A")
    player_b = Player("B")


    deck = Deck()
    
    # Shuffling the created deck
    
    deck.shuffle()

    for i in range(26):
        # Creating Player's decks from created Deck
        
        player_a.all_card.append(deck.deal_one())
        player_b.all_card.append(deck.deal_one())


    while game_on:

        round += 1
        print(f"Round {round}.")
        
        # Defining Win conditions, checking if it's true and end the game if it is the case
        
        won_a = len(player_b.all_card) == 0
        won_b = len(player_a.all_card) == 0
        
        
        if won_a and won_b:
            print("DRAW!")
        elif won_a:
            print("Player A Won!")
            print("Game Over!")
            game_on = False
            continue
        elif won_b:
            print("Player B Won!")
            print("Game Over!")
            game_on = False
            continue
        else:
            pass
        
        
        # Played cards of both players to be stored here
        
        a_cards = []
        b_cards = []
        
        
        # Making the play of card from both players' end
        
        a_cards.append(player_a.remove_card())
        b_cards.append(player_b.remove_card())

        # To compare the value assuming the war condition and denying it if found to be the case
        """
        It's assumed previously to deal with the wars that might stack upon each other.
        """
        
        at_war = True
        while at_war:
            
            print(f"""
            Player A's card - {a_cards[-1]}
            Player B's card - {b_cards[-1]}""")
            
            if a_cards[-1].value > b_cards[-1].value:

                player_a.add_card(a_cards)
                player_a.add_card(b_cards)
                print(f"""
                {player_a}
                {player_b}""")
                break
            elif a_cards[-1].value < b_cards[-1].value:

                player_b.add_card(a_cards)
                player_b.add_card(b_cards)
                print(f"""
                {player_a}
                {player_b}""")
                break
            else:
                print("WAR BEGINS!")
                
                # To check if players are even able to play the war
                
                if len(player_a.all_card) < 10:
                    print("Player A unable to wage war.")
                    print("Player B Won!")
                    game_on = False
                    break
                elif len(player_b.all_card) < 10:
                    print("Player B unable to wage war.")
                    print("Player A Won!")
                    game_on = False
                    break
                else:
                    # If both players are able to pley war.
                    
                    for i in range(10):
                        a_cards.append(player_a.remove_card())
                        b_cards.append(player_b.remove_card())
                    continue
```
